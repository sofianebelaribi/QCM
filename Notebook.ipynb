```python
from docx import Document

def lire_qcm_de_word(chemin_fichier):
    doc = Document(chemin_fichier)
    questions = []
    question_en_cours = None

    for paragraphe in doc.paragraphs:
        if paragraphe.text.startswith("QCM sur word"):
            continue  # Ignorer l'en-tête

        if paragraphe.text.startswith("1)") or paragraphe.text.startswith("2)") or \
                paragraphe.text.startswith("3)") or paragraphe.text.startswith("4)") or \
                paragraphe.text.startswith("5)"):
            if question_en_cours:
                questions.append(question_en_cours)
            question_en_cours = {"question": paragraphe.text[2:], "reponses": []}
        elif paragraphe.text.startswith("A.") or paragraphe.text.startswith("B.") or \
                paragraphe.text.startswith("C.") or paragraphe.text.startswith("D.") or \
                paragraphe.text.startswith("E.") or paragraphe.text.startswith("F.") or \
                paragraphe.text.startswith("G.") or paragraphe.text.startswith("H.") or \
                paragraphe.text.startswith("I.") or paragraphe.text.startswith("J."):
            est_bonne_reponse = paragraphe.runs[0].bold  # Vérifie si le texte est en gras
            reponse = f"V:{paragraphe.text[2:]}" if est_bonne_reponse else paragraphe.text[2:]
            question_en_cours["reponses"].append(reponse)

    if question_en_cours:
        questions.append(question_en_cours)

    return questions

def convertir_en_syntaxe_texte(questions):
    for i, qcm in enumerate(questions, 1):
        question = qcm["question"]
        reponses = " | ".join(qcm["reponses"])
        print(f"QCM || Question {i} || {reponses}")

if __name__ == "__main__":
    chemin_fichier_word = "./QCM1.docx"
    qcm = lire_qcm_de_word(chemin_fichier_word)
    convertir_en_syntaxe_texte(qcm)
```

```python
from docx import Document

def lire_qcm_de_word(chemin_fichier):
    doc = Document(chemin_fichier)
    questions = []
    question_en_cours = None

    for paragraphe in doc.paragraphs:
        lignes = paragraphe.text.split('\n')

        for ligne in lignes:
            if ligne.strip():  # Ignorer les lignes vides
                if ligne.startswith("QCM sur word"):
                    continue  # Ignorer l'en-tête

                if any(ligne.startswith(f"{i})") for i in range(1, 1001)):
                    if question_en_cours:
                        questions.append(question_en_cours)
                    question_en_cours = {"question": ligne[ligne.find(")") + 1:].strip(), "reponses": []}
                elif ligne.startswith("A.") or ligne.startswith("B.") or \
                        ligne.startswith("C.") or ligne.startswith("D.") or \
                        ligne.startswith("E.") or ligne.startswith("F.") or \
                        ligne.startswith("G.") or ligne.startswith("H.") or \
                        ligne.startswith("I.") or ligne.startswith("J."):
                    est_bonne_reponse = ligne[2:].strip().startswith("V:")
                    reponse = f"V:{ligne[2:]}" if est_bonne_reponse else ligne[2:]
                    question_en_cours["reponses"].append(reponse)

    if question_en_cours:
        questions.append(question_en_cours)

    return questions

def convertir_en_syntaxe_texte(questions):
    for i, qcm in enumerate(questions, 1):
        question = qcm["question"]
        reponses = " | ".join(qcm["reponses"])
        print(f"QCM || Question {i} || {reponses}")

if __name__ == "__main__":
    chemin_fichier_word = "./QCM1.docx"
    qcm = lire_qcm_de_word(chemin_fichier_word)
    convertir_en_syntaxe_texte(qcm)
```
